```python
#Problem_Set_1_Q4.py
""" 
Problem_Set_1 Question 4:
Explain how you can use Decision Trees to perform regression? Show that when the error function
is squared error, then the expected value at any leaf is the mean. Take the Boston Housing dataset
(https://archive.ics.uci.edu/ml/datasets/Housing) and use Decision Trees to perform regression.

Starting point for solution:
https://towardsdatascience.com/https-medium-com-lorrli-classification-and-regression-analysis-with-decision-trees-c43cdbc58054
"""
import pandas as pd
from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

class BostonDataColumn:
    def __init__(self, num, title, descr, max_depth):
        self.num = num
        self.title = title
        self.descr = descr
        self.max_depth = max_depth

def getColumns():
    columns = []
    column = BostonDataColumn(0, "CRIM",   "Per capita crime rate by town",3)
    columns.append(column)
    column = BostonDataColumn(1, "ZN",     "Proportion of residential land zoned for lots over 25,000 sq.ft.",2)
    columns.append(column)
    column = BostonDataColumn(2, "INDUS",  "Proportion of non-retail business acres per town",2)
    columns.append(column)
    # Since all these are 1 or 0, not much use in this column
    #column = BostonDataColumn(3, "CHAS",   "Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)",3)
    #columns.append(column)
    column = BostonDataColumn(4, "NOX",    "Nitric oxides concentration (parts per 10 million)",2)
    columns.append(column)
    column = BostonDataColumn(5, "RM",     "Average number of rooms per dwelling",2)
    columns.append(column)
    column = BostonDataColumn(6, "AGE",    "Proportion of owner-occupied units built prior to 1940",2)
    columns.append(column)
    column = BostonDataColumn(7, "DIS",    "Weighted distances to five Boston employment centres",2)
    columns.append(column)
    column = BostonDataColumn(8, "RAD",    "Index of accessibility to radial highways",3)
    columns.append(column)
    column = BostonDataColumn(9, "TAX",    "Full-value property-tax rate per $10,000",2)
    columns.append(column)
    column = BostonDataColumn(10,"PTRATIO","Pupil-teacher ratio by town",2)
    columns.append(column)
    column = BostonDataColumn(11,"B",      "1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town",2)
    columns.append(column)
    column = BostonDataColumn(12,"LSTAT",  "% lower status of the population",3)
    columns.append(column)
    return columns
        
def graphDecisionTreeRegressorAtColumn(column, column_name, x_axis, max_depth):
    # Create DataFrame using only the chosen feature
    df = pd.DataFrame(boston.data[:, column])
    df.columns = [column_name]
    
    # Create new column for the target MEDV
    target_name = 'MEDV'
    df[target_name] = boston.target
    
    # If you want to see some of the data:
    #print(df.head())
    
    # Matrix X
    X = df[[column_name]].values                      
    # Vector y
    y = df[target_name].values                             
    
    # Sort X and y by ascending values of X
    sort_idx = X.flatten().argsort()                  
    X = X[sort_idx]
    y = y[sort_idx]
    
    # Initialize and fit regressor using (Weighted) Mean Squared Error
    # max_depth: tricky to find optimal balance between overfitting and underfitting for each column
    tree = DecisionTreeRegressor(criterion='mse', max_depth=max_depth, random_state=0)
    tree.fit(X, y)
    
    plt.figure(figsize=(16, 8))
    # Plot actual target against features
    plt.scatter(X, y, c='steelblue', edgecolor='white', s=70)
    # Plot predicted target against features
    plt.plot(X, tree.predict(X), color='black', lw=2)
    plt.xlabel(x_axis + ' ['+column_name+']')
    plt.ylabel('Price in $1000s ['+target_name+']')
    plt.show()
    
# Load Boston Dataset
boston = datasets.load_boston()            
#print("dataset description:")
#print(boston.DESCR) 
columns = getColumns()
for column in columns:
    graphDecisionTreeRegressorAtColumn(column.num, column.title, column.descr, column.max_depth)
```
